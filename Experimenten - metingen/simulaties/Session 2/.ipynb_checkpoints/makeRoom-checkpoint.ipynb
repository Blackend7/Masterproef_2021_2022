```python
def makeRoom(room, wall_perm, door_perm):

    wt = 2
    it = 2
    dt = 6

    room = room - 1
    wall_perm = wall_perm - 1
    door_perm = door_perm - 1

    perm_ratio = door_perm / wall_perm

    #perimeter
    room[:it, :] = wall_perm
    room[-it:, :] = wall_perm
    room[:, :it] = wall_perm
    room[:, -it:] = wall_perm

    #hall
    nx, ny = np.shape(room)
    nx4_10 = round(4*nx/10)
    nx6_10 = round(6*nx/10)

    room[:, nx4_10-1:nx4_10+wt-2] = wall_perm
    room[:, nx6_10-1:nx6_10+wt-2] = wall_perm

    #left rooms
    ny2_3 = round(2*ny/3);

    room[ny2_3:ny2_3+wt-1, 1:nx4_10] = wall_perm

    #right rooms
    ny1_4 = round(1*ny/4)
    ny3_4 = round(3*ny/4)

    room[ny1_4-1:ny1_4+wt-2, nx6_10+wt-2:] = wall_perm
    room[ny3_4-1:ny3_4+wt-2, nx6_10+wt-2:] = wall_perm

    #attribute doors
    nx1_2 = round(nx/2)
    ny9_10 = round(9*ny/10)
    ny1_10 = round(ny/10)

    #to hal
    room[:it, nx1_2-dt:nx1_2+dt-1] = room[:it, nx1_2-dt:nx1_2+dt-1]*perm_ratio

    #to top rooms
    room[ny1_10-dt+10:ny1_10+dt+9, wt:-wt] = room[ny1_10-dt+10:ny1_10+dt+10-1, wt:-wt]*perm_ratio

    #to bottom rooms
    room[ny9_10-dt-10:ny9_10+dt-9, wt:-wt] = room[ny9_10-dt-10:ny9_10+dt-9, wt:-wt]*perm_ratio

    #to middle right room
    room[nx1_2-dt:nx1_2+dt-1, nx4_10-1:-wt] = room[nx1_2-dt:nx1_2+dt-1, nx4_10-1:-wt]*perm_ratio

    room = room + 1
    return room
```
